```python
import subprocess

GITHUB_USER  = "thejorseman"
GITHUB_TOKEN = "ghp_xxxxxxxxxxxxxxxxxxxx"   # Settings > Developer settings > Personal access tokens
GITHUB_REPO  = "CloneCharter"
BRANCH       = "main"

remote_url = f"https://{GITHUB_USER}:{GITHUB_TOKEN}@github.com/{GITHUB_USER}/{GITHUB_REPO}.git"

cmds = [
    ["git", "add", "-A"],
    ["git", "commit", "--allow-empty", "-m", "chore: update scripts (push_model, gradio_multigen)"],
    # -c credential.helper='' desactiva el credential manager solo para este push
    # -> git usa el token embebido en la URL en lugar de las credenciales cacheadas
    ["git", "-c", "credential.helper=", "push", remote_url, f"HEAD:{BRANCH}"],
]

for cmd in cmds:
    display_cmd = " ".join(cmd).replace(GITHUB_TOKEN, "***")
    print(f"$ {display_cmd}")
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.stdout:
        print(result.stdout.strip())
    if result.returncode != 0:
        print(f"ERROR: {result.stderr.strip()}")
        break
else:
    print(f"Push completado: github.com/{GITHUB_USER}/{GITHUB_REPO} ({BRANCH})")
```
